```python
import os
import openai
from dotenv import load_dotenv
# Load API key from environment variable
load_dotenv(override=True)
openai_key = os.getenv("OPENAI_API_KEY")
client = openai.OpenAI(api_key=openai_key)
```

```python
# System message defines personality/behavior
messages = [{"role": "system", "content": "You are a friendly tutor who explains simply."}]

while True:
    user_input = input("You: ")

    # Stop the chat
    if user_input.lower() in ["exit", "quit"]:
        break
    # Add user's message to the conversation
    messages.append({"role": "user", "content": user_input})

    # Send conversation history to the model
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages
    )

    # Get model's reply
    assistant_reply = response.choices[0].message.content
    print("Bot:", assistant_reply)

    # Add assistant's reply to the history
    messages.append({"role": "assistant", "content": assistant_reply})
```

```python
def get_completion(messages, model="gpt-3.5-turbo", temperature=0, max_tokens=150):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, # Degree of randomness of the output
         max_tokens=max_tokens)
    content = response.choices[0].message.content

    token_dict = {
        'prompt_tokens': response.usage.prompt_tokens,
        'completion_tokens': response.usage.completion_tokens,
        'total_tokens': response.usage.total_tokens,}

    return content, token_dict
```

```python
messages = [{'role':'system', 'content':"You are a history teacher in a primary school"},
{'role':'user', 'content':"What is the capital of France?"}]
```

```python
response, token_dict = get_completion (messages)
print (response)
print (token_dict)
```

```python
total_cost = ((token_dict['prompt_tokens'] * 0.0005) / 1000) + ((token_dict['completion_tokens'] * 0.0015) / 1000)
print(f"The total cost = ${total_cost:.6f}")
```

```python
# Streaming response
stream = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": "Write a long essay about AI"}],
    stream=True
)

for chunk in stream:
    if chunk.choices[0].delta.content:
        print(chunk.choices[0].delta.content, end="", flush=True)
```

```python
# Non-Streaming response
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": "Write a long essay about AI"}]
)
content = response.choices[0].message.content
print(content)
```
