```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

**Example text**

This is just a normal Python string that we will process using NLP tools.

```python
worf_quote = "Sir, I protest. I am not a merry man!"
```

**Tokenise the text into words**
This splits the sentence into words and punctuation, because punctuation can also carry meaning in NLP.

```python
words_in_quote = word_tokenize(worf_quote)
words_in_quote
['Sir', ',', 'protest', '.', 'merry', 'man', '!']
```

**Load English stop words**
Stop words are very common words like “I”, “am”, “a”, “the” that often add little topic meaning.

```python
stop_words = set(stopwords.words("english"))
```

**Create an empty list for filtered words**

This list will store only the words that are not stop words.

```python
filtered_list = []
```

Filter out stop words using a loop

casefold() converts words to lowercase so comparisons work correctly with the stop word list.

```python
for word in words_in_quote:
  if word.casefold() not in stop_words:
         filtered_list.append(word)
```

**Same filtering using list comprehension (shorter way)**

```python
filtered_list = [
    word for word in words_in_quote if word.casefold() not in stop_words
]
```

**Stemming**

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
stemmer = PorterStemmer()
```

```python
string_for_stemming = """
The crew of the USS Discovery discovered many discoveries.
Discovering is what explorers do."""
```

```python
words = word_tokenize(string_for_stemming)
```

```python
words
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
```
